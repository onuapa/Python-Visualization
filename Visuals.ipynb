```python
plt.figure(figsize=(13,5))
xt = np.array([0, 0.1, 0.2, 0.3, 0.5, 0.8, 1.0])
yt = np.array([1.0, 2.1, 3.5, 6.5, 7.2, 5.9, 6.3])
zt = xt * yt

plt.scatter(xt, yt, c=zt, marker='s')
plt.scatter(xt, yt, c=zt, marker='^')
plt.xlabel('xt')
plt.ylabel('yt')
plt.colorbar(label='zt')
plt.title('Scatter Plot of yt and zt')
plt.legend(['yt', 'zt'])
plt.show()
plt.savefig("josephkyeremeh_scatter.png")
```

```python
# Generate sample data
np.random.seed(42)
data = np.random.randn(100, 2)
categories = ['Category A', 'Category B', 'Category C', 'Category D']
category_counts = [30, 15, 25, 30]
categorical_data = np.random.choice(categories, size=100, p=category_counts / np.sum(category_counts))
numeric_data = np.random.randint(1, 6, size=100)
heatmap_data = np.random.rand(10, 10)

# Create subplots
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle('Dashboard of Learnt Visuals', fontsize=16)

# Pie Chart
axes[0, 0].pie(category_counts, labels=categories, autopct='%1.1f%%', startangle=90)
axes[0, 0].set_title('Pie Chart')

# Scatter Plot
axes[0, 1].scatter(data[:, 0], data[:, 1])
axes[0, 1].set_title('Scatter Plot')

# Bar Plot
unique, counts = np.unique(categorical_data, return_counts=True)
axes[0, 2].bar(unique, counts)
axes[0, 2].set_xticks(unique)
axes[0, 2].set_title('Bar Plot')

# Histogram
axes[1, 0].hist(numeric_data, bins=5, edgecolor='black')
axes[1, 0].set_title('Histogram')

# Heatmap
sns.heatmap(heatmap_data, cmap='viridis', ax=axes[1, 1])
axes[1, 1].set_title('Heatmap')

# Remove empty subplot
fig.delaxes(axes[1, 2])

# Adjust layout and show the dashboard
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
